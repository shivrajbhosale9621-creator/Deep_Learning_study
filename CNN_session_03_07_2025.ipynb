```python
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
#Notmalize the images

X_train_normalized=X_train/255
X_test_normalized=X_test/255
```

```python
plt.figure(figsize=(16,16))
for i in range(1,26):
  plt.subplot(5,5,i)
  plt.imshow(X_train[i-1])
  plt.colorbar()
  plt.title(class_names[y_train[i-1][0]])
```

```python
from tensorflow.keras import datasets,layers,models
model=models.Sequential()
#################################################3

model.add(layers.Conv2D(32,#Filters
                       (3,3), # Filter size/kernal size
                       activation='relu',
                       input_shape=(32,32,3)
                       )) # Stride  / pading :Defalut

model.add(layers.MaxPool2D(2,2)) # shape of the pool

#########################################################################
model.add(layers.Conv2D(64,#Filters
                        (3,3), # Filters size/kernal size
                        activation='relu'))
model.add(layers.MaxPool2D(2,2))

#########################################################################
model.add(layers.Conv2D(64,(3,3),activation='relu'))
```

```python
model.summary()
```

```python
# Step-5
# ADd dense laters

model.add(layers.Flatten()) # After
#no need of input shape here
model.add(layers.Dense(64,activation='relu')) # Hidden layer 64 Nurons,relu activation function
#model.add(layers.Dense(10)) 10 classes so 10 neurins
model.add(layers.Dense(10,activation='softmax')) # Softmax activation fucntion
```

```python
model.summary()
```

```python
#Complie
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val=train_test_split(X_train,y_train,test_size=0.2,random_state=41)
```

```python
history=model.fit(X_train,
                  y_train,
                  epochs=25,
                  batch_size=64,
                  validation_data=(X_val,y_val))
```

```python
history.history
```

```python
history.history.keys()
```

```python
plt.plot(history.history['accuracy'],label='accuracy')
plt.plot(history.history['val_accuracy'],label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
```

```python
y_pred=model.predict(X_test)
```

```python
plt.figure(figsize=(20,25))
for i in range(100):
  plt.subplot(10,10,i+1)
  plt.imshow(X_test[i])
  actual=class_names[y_test[i][0]]
  predicted=class_names[np.argmax(y_pred[i])]
  plt.title(f"Actual:{actual},Predicted:{predicted}")
```
