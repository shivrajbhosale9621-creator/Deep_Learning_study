```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Resize image (assuming 'img' is already defined)
img_resized = cv2.resize(X_train[0], (244, 244), interpolation=cv2.INTER_CUBIC)

# Step 2: Convert to float32 for better filtering
img_float = img_resized.astype(np.float32)

# Step 3: Apply sharpening using filter2D
sharp_kernel = np.array([
    [-1, -1, -1],
    [-1,  9, -1],
    [-1, -1, -1]
], dtype=np.float32)

sharp_f2d = cv2.filter2D(img_float, ddepth=-1, kernel=sharp_kernel)
sharp_f2d_norm = cv2.normalize(sharp_f2d, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

# Step 4: Apply sharpening using Unsharp Masking
blurred = cv2.GaussianBlur(img_float, (5, 5), 0)
unsharp = cv2.addWeighted(img_float, 1.5, blurred, -0.5, 0)
unsharp_norm = cv2.normalize(unsharp, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

# Step 5: Plot the results
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.title("Original")
plt.imshow(img_resized, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Sharpened (filter2D)")
plt.imshow(sharp_f2d_norm, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Unsharp Masking")
plt.imshow(unsharp_norm, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Step-8:Compile

```python
#sir code

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.initializers import GlorotUniform  # Xavier initialization
from tensorflow.keras.activations import relu, softmax

# Define the model
model = Sequential()

# Input layer: Flatten 28x28 to 784
model.add(Flatten(input_shape=(28, 28)))

# Hidden layer
model.add(Dense(128, activation='relu', kernel_initializer=GlorotUniform()))  # Xavier Init
model.add(BatchNormalization())  # Normalize activations
model.add(Dropout(0.3))  # Dropout for regularization (30% dropped)

# Output layer
model.add(Dense(10, activation='softmax', kernel_initializer=GlorotUniform()))  # Xavier Init

# Print model summary
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',  # ✅ corrected spelling
              metrics=['accuracy'])
```

Step-9:Divide train data into train validation

```python
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val=train_test_split(X_train,Y_train,test_size=0.2,random_state=42)
```

# Model fitting

```python
history=model.fit(X_train,
                  y_train,
                  epochs=10,
                  batch_size=64,
                  validation_data=(X_val,y_val))
```
