```python
import cv2
```

```python
vid=cv2.VideoCapture(r"C:\Users\Shivraj\Downloads\WhatsApp Video 2025-09-02 at 6.38.24 PM.mp4")

while True:
    # capture the video frame
    # by frame
    ret,frame=vid.read()
    #ret=boolean value
    #ret=True ,the frame is captured
    #ret=False the frame is not capture
    print(ret)
    #Display the resulting frame
    cv2.imshow('frame',frame)
    #the 'q' button is set as the 
    #quitting button you may use any
    #desired button of your choice
    # 1 means one milli second
    #ord('q'):ASCII 8 digits
    #so it will check are you entered any 8 digit key or not
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break

vid.release()
cv2.destroyAllWindows()
```

```python
import os
vid=cv2.VideoCapture(r"C:\Users\Shivraj\Downloads\WhatsApp Video 2025-09-02 at 6.38.24 PM.mp4")
try:
    #creating a floder named data
    if not os.path.exists('data'):
        os.makedirs('data')
except OSError:
    print('Error:Creating directory of data')
currentframe=0
while True:
    ret,frame=vid.read()
    if ret:
        name='./data/frames'+str(currentframe)+'.jpg' #inside data
        print('Creating...'+name)               
        #writing the extracted images
        cv2.imwrite(name,frame)
        cv2.imshow('frame',frame)
        currentframe+=1
    else:
        break
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break


vid.release()
cv2.destroyAllWindows()
```
